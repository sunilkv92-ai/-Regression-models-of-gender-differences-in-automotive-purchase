```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
df = pd.read_excel("cleaned_car1.xlsx")
```

```python
#df
```

```python
df.Displacement = [x.replace(" cm3", "") for x in df.Displacement]
```

```python
import re
def remove_suffix(value):
    return re.sub(r'\s*\(.*\)|\sin', '', value)
df.GroundClearance = [remove_suffix(value) for value in df.GroundClearance]
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[['GroundClearance']],df['Male'], test_size=0.25)
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
r2 = r2_score(y_test, y_pred)
```

```python
plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, model.predict(X_train), color='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression')
plt.show()
```

```python
X_train = X_train['GroundClearance'].values.ravel()
y_train = y_train.values.ravel()

# Plot the training data and the linear regression line
plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, model.predict(X_train.reshape(-1, 1)), color='blue')  # Reshape X_train to a 2D array
plt.xlabel('GroundClearance')
plt.ylabel('Male')
plt.title('Linear Regression')
plt.show()
```
