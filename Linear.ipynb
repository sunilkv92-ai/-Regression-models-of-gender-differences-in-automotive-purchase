```python
import pandas as pd
import re

# Load the dataset into a pandas DataFrame
df = pd.read_excel("cleaned_car1.xlsx")

# Define a function to extract the numeric value before "mpg"
def extract_highway_value(value):
    try:
        # Use regular expression to find the numeric part before "mpg"
        match = re.search(r'(\d+)', value)
        if match:
            return int(match.group(0))
        else:
            return None
    except ValueError:
        return None

# Extract the numeric value before "mpg" from the 'Highway' column
df['Highway'] = df['Highway'].apply(extract_highway_value)
```
