```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import MinMaxScaler


df = pd.read_excel("project_car_data.xlsx")


X = df[['Highway', 'Displacement', 'Tiresize', 'Wheelbase', 'GroundClearance', 'Cargo', 'Weight']]
y_male = df['Male']
y_female = df['Female']


scaler = MinMaxScaler()
X_normalized = scaler.fit_transform(X)

X_train_male, X_test_male, y_train_male, y_test_male = train_test_split(X_normalized, y_male, test_size=0.2, random_state=42)
X_train_female, X_test_female, y_train_female, y_test_female = train_test_split(X_normalized, y_female, test_size=0.2, random_state=42)

regressor_male = RandomForestRegressor(n_estimators=100, random_state=42)
regressor_male.fit(X_train_male, y_train_male)

regressor_female = RandomForestRegressor(n_estimators=100, random_state=42)
regressor_female.fit(X_train_female, y_train_female)


y_pred_male = regressor_male.predict(X_test_male)
y_pred_female = regressor_female.predict(X_test_female)



print("Male - Mean Squared Error:", mean_squared_error(y_test_male, y_pred_male))
print("Male - R-squared:", r2_score(y_test_male, y_pred_male))

print("Female - Mean Squared Error:", mean_squared_error(y_test_female, y_pred_female))
print("Female - R-squared:", r2_score(y_test_female, y_pred_female))


```

```python
import matplotlib.pyplot as plt

# Scatter plot for Male Model
plt.figure(figsize=(8, 6))
plt.scatter(y_test_male, y_pred_male, color='blue', alpha=0.5)
plt.plot([min(y_test_male), max(y_test_male)], [min(y_test_male), max(y_test_male)], color='black', linestyle='--', linewidth=2)
plt.xlabel("True Values (Male)")
plt.ylabel("Predictions (Male)")
plt.title("Male Model: True Values vs. Predictions")
plt.grid(True)
plt.show()


```

```python
# Scatter plot for Female Model
plt.figure(figsize=(8, 6))
plt.scatter(y_test_female, y_pred_female, color='red', alpha=0.5)
plt.plot([min(y_test_female), max(y_test_female)], [min(y_test_female), max(y_test_female)], color='black', linestyle='--', linewidth=2)
plt.xlabel("True Values (Female)")
plt.ylabel("Predictions (Female)")
plt.title("Female Model: True Values vs. Predictions")
plt.grid(True)
plt.show()
```
