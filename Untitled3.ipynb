```python
import pandas as pd
import re

# Load the dataset into a pandas DataFrame
df = pd.read_excel("cleaned_car1.xlsx")

# Define a function to extract the value before "mpg"
def extract_value_before_mpg(value):
    try:
        # Use regular expression to find the numeric part before "mpg"
        match = re.search(r'(\d+)', value)
        if match:
            return int(match.group(0))
        else:
            return None
    except ValueError:
        return None

# Extract the numeric value before "mpg" from the 'Highway' column
df['Highway'] = df['Highway'].apply(extract_value_before_mpg)
```

```python
df
```

```python
# Define a function to remove "cm3" from the values in the 'Displacement' column
def remove_cm3(value):
    try:
        # Remove 'cm3' from the string and convert to integer
        return int(value.replace("cm3", "").strip())
    except ValueError:
        return None

# Remove "cm3" from the 'Displacement' column
df['Displacement'] = df['Displacement'].apply(remove_cm3)
```

```python
df
```

```python
# Define a function to keep the data before '/'
def keep_data_before_slash(value):
    try:
        # Find the part before '/' and remove leading/trailing whitespaces
        return value.split('/')[0].strip()
    except AttributeError:
        return None

# Keep data before '/' from the 'Tiresize' column
df['Tiresize'] = df['Tiresize'].apply(keep_data_before_slash)
```

```python
df
```

```python
# Define a function to keep the data before 'in'
def keep_data_before_in(value):
    try:
        # Find the part before 'in' and remove leading/trailing whitespaces
        return value.split('in')[0].strip()
    except AttributeError:
        return None

# Keep data before 'in' from the 'Wheelbase' and 'GroundClearance' columns
df['Wheelbase'] = df['Wheelbase'].apply(keep_data_before_in)
df['GroundClearance'] = df['GroundClearance'].apply(keep_data_before_in)
```

```python
df
```

```python
# Define a function to keep the data before 'K' and 'lbs'
def keep_data_before_unit(value, unit):
    try:
        # Find the part before the unit and remove leading/trailing whitespaces
        return value.split(unit)[0].strip()
    except AttributeError:
        return None

# Keep data before 'K' from the 'Cargo' column
df['Cargo'] = df['Cargo'].apply(keep_data_before_unit, args=('c',))

# Keep data before 'lbs' from the 'Weight' column
df['Weight'] = df['Weight'].apply(keep_data_before_unit, args=('lbs',))
```

```python
df
```

```python
df.to_excel("final_car_data.xlsx", index=False)
```

```python
import pandas as pd

# Load the dataset into a pandas DataFrame
df = pd.read_excel("final_car_data.xlsx")

# Define a function to keep the first integer value from the 'Tiresize' column
def keep_first_integer(value):
    try:
        # Find the first integer in the string and return it
        return int(''.join(filter(str.isdigit, str(value))))
    except ValueError:
        return None

# Keep the first integer value from the 'Tiresize' column
df['Tiresize'] = df['Tiresize'].apply(keep_first_integer)
```
